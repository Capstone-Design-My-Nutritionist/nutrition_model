```python
import numpy as np

# 1) 빈 문자열(공백 포함)을 NaN으로 변환
df_clean = df_clean.replace(r'^\s*$', np.nan, regex=True)

# 2) 모든 결측치(NaN)를 0으로 채우기
df_clean = df_clean.fillna(0)

# 결과 확인
print(df_clean.columns.tolist())
```

```python
# 1) 제거할 영양소 컬럼 리스트
drop_cols = ['수분(g)', '회분(g)', '인(mg)', '칼륨(mg)']

# 2) DataFrame에서 해당 컬럼 제거 (없는 컬럼은 무시)
df_clean = df_clean.drop(columns=drop_cols, errors='ignore')

# 3) 남은 컬럼 확인
print("남은 컬럼:", df_clean.columns.tolist())

# 4) head()로 샘플 보기
print(df_clean.head())
```

```python
# 1) B-complex 단순 합산 (mg 단위)
b_cols = ['티아민(mg)', '리보플라빈(mg)', '니아신(mg)']
df_clean['vitamin_B_total_mg'] = df_clean[b_cols].sum(axis=1)

# 2) RDA 기반 가중 결합 스코어 계산
RDA_B = {
    '티아민(mg)':     1.1,   # 예시 RDA 값 (mg)
    '리보플라빈(mg)': 1.2,
    '니아신(mg)':    14.0
}

def compute_b_score(row):
    vals = []
    for col, rda in RDA_B.items():
        frac = row[col] / rda
        vals.append(min(1, max(0, frac)))  # 0~1 로 클립
    return sum(vals) / len(vals)

df_clean['vitamin_B_score'] = df_clean.apply(compute_b_score, axis=1)

# 3) 결과 예시 확인
df_clean[['티아민(mg)', '리보플라빈(mg)', '니아신(mg)',
          'vitamin_B_total_mg', 'vitamin_B_score']].head(10)
```
